```python
import numpy as np
from scipy.linalg import eigh
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Mass matrix (3-storey building example)
m = np.array([[18438.6, 0, 0],
              [0, 13761, 0],
              [0, 0, 9174]])

# Damping matrix
c = 100000 * np.array([[0.82836, -0.6037, 0],
                       [-0.6037, 1.4821, -0.7824],
                       [0, -0.7824, 0.84647]])

# Stiffness matrix
k = np.array([[74504937, -64228394, 0],
              [-64228394, 147468394, -83240000],
              [0, -83240000, 83240000]])

# Force parameters
F0 = 58000  # Peak force
time = np.linspace(0, 10, 1000)  # Time vector (0 to 10 seconds, 1000 points)
f_ext = F0 * np.sin(2 * np.pi * 1 * time)  # Sinusoidal force on the top level

# Eigenvalue problem (modal analysis)
eigenvalues, eigenvectors = eigh(k, m)  # Eigenvalue decomposition

# Natural frequencies
omega_n = np.sqrt(eigenvalues)  # Natural frequencies in rad/s
frequencies = omega_n / (2 * np.pi)  # Convert to Hz

print("Natural Frequencies (Hz):", frequencies)

# Time History Analysis
def time_history_analysis(state, t, m, c, k, f_ext):
    # System of 1st order differential equations
    u = state[:3]  # Displacements
    v = state[3:]  # Velocities

    f = np.array([0, 0, f_ext(t)])  # External force applied at top level
    a = np.linalg.inv(m).dot(f - c.dot(v) - k.dot(u))  # Accelerations

    return np.concatenate([v, a])

# Initial conditions (displacement and velocity = 0)
initial_conditions = np.zeros(6)

# Solve time history analysis using odeint
solution = odeint(time_history_analysis, initial_conditions, time, args=(m, c, k, lambda t: F0 * np.sin(2 * np.pi * 1 * t)))

# Extract displacement at the top level (u3)
u3 = solution[:, 2]

# Maximum displacement
u3_max_time_history = np.max(np.abs(u3))

print("Maximum displacement (u3) from Time History Analysis:", u3_max_time_history)

# Plot the displacement response over time
plt.figure()
plt.plot(time, u3, label='u3 (Top Level Displacement)')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Time History Response of Top Level (u3)')
plt.legend()
plt.grid(False)
plt.show()

# (b) Response Spectrum Method

# Response spectrum (example: assuming response values for each mode)
# Here we'll use an example array for demonstration. Replace this with your own response spectrum data.
response_spectrum = np.array([0.1, 0.05, 0.02])  # Example spectral displacement for each mode

# Modal participation factors (assuming unit participation for simplicity)
modal_participation = np.ones(3)

# Maximum response per mode
modal_max = response_spectrum * modal_participation

# Combine modal maxima using SRSS method
u3_max_response_spectrum = np.sqrt(np.sum(modal_max ** 2))

print("Maximum displacement (u3) from Response Spectrum Method:", u3_max_response_spectrum)

# SRSS Combination
u3_max_srss = np.sqrt(u3_max_time_history**2 + u3_max_response_spectrum**2)
print("Combined Maximum Displacement (SRSS):", u3_max_srss)
```
