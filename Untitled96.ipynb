```python
import numpy as np
import matplotlib.pyplot as plt

# System parameters
m = 100  # Mass (kg)
k = 100000  # Stiffness (N/m)
zeta = 0.1  # Damping ratio
omega_n = np.sqrt(k / m)  # Natural frequency (rad/s)
c = 2 * zeta * m * omega_n  # Damping coefficient (Ns/m)

# Forcing function parameters
T = 0.12  # Period of the forcing function (s)
omega_0 = 2 * np.pi / T  # Fundamental frequency (rad/s)
F0 = 1000  # Amplitude of the force (N), assumed value

# Number of Fourier terms
N_terms = 20

# Time array for plotting
t = np.linspace(0, T, 1000)

# Initialize Fourier coefficients
a0 = 2 * F0
an = np.zeros(N_terms)
bn = np.zeros(N_terms)

# Compute Fourier coefficients for the force
for n in range(1, N_terms + 1):
    if n % 2 == 1:  # Odd n
        bn[n-1] = (4 * F0) / (n * np.pi)
    else:  # Even n
        bn[n-1] = 0

# Initialize displacement response
x = np.zeros_like(t)

# Compute displacement response using Fourier series
for n in range(1, N_terms + 1):
    omega = n * omega_0
    H = 1 / (k - m * omega**2 + 1j * c * omega)  # Transfer function
    An = np.real(H) * 0 - np.imag(H) * bn[n-1]  # a_n = 0 for n > 0
    Bn = np.real(H) * bn[n-1] + np.imag(H) * 0  # a_n = 0 for n > 0
    x += An * np.cos(n * omega_0 * t) + Bn * np.sin(n * omega_0 * t)

# Add the DC term (a0/2)
x += a0 / 2 * np.real(1 / (k + 1j * c * 0))  # DC term

# Plot the steady-state response
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='Steady-State Response')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Steady-State Displacement Response of SDOF System')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Constants and parameters
Q = 10  # Emission rate (g/s)
d = 1000  # Distance from source to receptor (m)
Zr = 50  # Receptor elevation (m)
u = 2  # Wind speed (m/s)
sigma_z = 20  # Vertical standard deviation of plume (m)
he = 100  # Effective plume height (m)
theta_0 = 0  # Mean wind direction (degrees)
sigma_theta = 15  # Standard deviation of wind direction (degrees)

# Convert wind direction standard deviation to radians
sigma_theta_rad = np.radians(sigma_theta)

# Gaussian plume model for concentration at a receptor
def gaussian_plume(theta):
    # Convert theta to radians
    theta_rad = np.radians(theta)

    # Calculate concentration using Gaussian plume model
    term1 = np.exp(-0.5 * ((Zr - he) / sigma_z) ** 2)
    term2 = np.exp(-0.5 * ((Zr + he) / sigma_z) ** 2)
    C = (Q / (2 * np.pi * u * sigma_z)) * (term1 + term2)

    return C

# Gaussian PDF for wind direction
def wind_direction_pdf(theta):
    theta_rad = np.radians(theta)
    theta_0_rad = np.radians(theta_0)
    pdf = (1 / (np.sqrt(2 * np.pi) * sigma_theta_rad)) * np.exp(-0.5 * ((theta_rad - theta_0_rad) / sigma_theta_rad) ** 2)
    return pdf

# Integrate the product of plume concentration and wind direction PDF
def integrand(theta):
    return gaussian_plume(theta) * wind_direction_pdf(theta)

# Perform numerical integration over all wind directions
total_concentration, _ = quad(integrand, -180, 180)

# Output the result
print(f"Total concentration at receptor: {total_concentration:.4f} g/m")

# Optional: Plot the wind direction PDF
theta_values = np.linspace(-180, 180, 360)
pdf_values = [wind_direction_pdf(theta) for theta in theta_values]

plt.plot(theta_values, pdf_values, label="Wind Direction PDF")
plt.xlabel("Wind Direction (degrees)")
plt.ylabel("Probability Density")
plt.title("Gaussian PDF of Wind Direction")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants and parameters
Q = 10  # Emission rate (g/s)
u = 2  # Wind speed (m/s)
sigma_y = 50  # Horizontal standard deviation of plume (m)
sigma_z = 20  # Vertical standard deviation of plume (m)
he = 100  # Effective plume height (m)

# Define a 2D grid for crosswind (x) and vertical height (z) at a fixed y
x = np.linspace(-200, 200, 100)  # Crosswind distance (m)
z = np.linspace(0, 200, 100)  # Vertical height (m)
X, Z = np.meshgrid(x, z)
y_fixed = 500  # Downwind distance (m) fixed

# Gaussian plume model function
def gaussian_plume(x, y, z):
    term1 = np.exp(-0.5 * (x ** 2) / (sigma_y ** 2))  # Horizontal spread
    term2 = np.exp(-0.5 * ((z - he) ** 2) / (sigma_z ** 2))  # Vertical spread (centerline)
    term3 = np.exp(-0.5 * ((z + he) ** 2) / (sigma_z ** 2))  # Reflection from ground
    C = (Q / (2 * np.pi * u * sigma_y * sigma_z)) * term1 * (term2 + term3)
    return C

# Compute concentration for fixed y = 500 m
C = gaussian_plume(X, y_fixed, Z)

# Plot the plume concentration as a contour plot
plt.figure(figsize=(10, 6))
contour = plt.contourf(X, Z, C, levels=50, cmap='viridis')
plt.colorbar(contour, label="Concentration (g/m)")
plt.xlabel("Crosswind Distance (m)")
plt.ylabel("Vertical Height (m)")
plt.title("Gaussian Plume Dispersion at y = 500 m")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Grid parameters
Nx, Ny, Nz = 50, 50, 30  # Number of grid points
Lx, Ly, Lz = 500, 1000, 300  # Domain size (meters)
dx, dy, dz = Lx / Nx, Ly / Ny, Lz / Nz  # Grid spacing
dt = 1  # Time step (seconds)
Nt = 500  # Number of time steps

# Constants
u = 5  # Wind speed in y-direction (m/s)
Kx, Ky, Kz = 10, 10, 5  # Eddy diffusivities (m/s)
k_SO2 = 0.0001  # SO oxidation rate (per second)
Q = 100  # Source emission rate (g/s)

# Initialize concentration field
C = np.zeros((Nx, Ny, Nz))  # SO concentration (g/m)

# Define source location
source_x, source_y, source_z = Nx // 2, 0, Nz // 2

# Time-stepping loop
for n in range(Nt):
    C_new = C.copy()

    # Apply finite difference scheme
    for i in range(1, Nx - 1):
        for j in range(1, Ny - 1):
            for k in range(1, Nz - 1):
                # Advection (upwind scheme)
                adv = -u * (C[i, j, k] - C[i, j - 1, k]) / dy

                # Diffusion (central difference scheme)
                diff_x = Kx * (C[i + 1, j, k] - 2 * C[i, j, k] + C[i - 1, j, k]) / dx**2
                diff_y = Ky * (C[i, j + 1, k] - 2 * C[i, j, k] + C[i, j - 1, k]) / dy**2
                diff_z = Kz * (C[i, j, k + 1] - 2 * C[i, j, k] + C[i, j, k - 1]) / dz**2

                # Chemical reaction (SO oxidation)
                reaction = -k_SO2 * C[i, j, k]

                # Update concentration using explicit scheme
                C_new[i, j, k] = C[i, j, k] + dt * (adv + diff_x + diff_y + diff_z + reaction)

    # Apply source at stack location
    C_new[source_x, source_y, source_z] += Q * dt / (dx * dy * dz)

    # Update concentration field
    C = C_new.copy()

# Visualization (Plot concentration at a fixed downwind slice)
Z, X = np.meshgrid(np.linspace(0, Lz, Nz), np.linspace(-Lx / 2, Lx / 2, Nx))  # Corrected meshgrid
C_slice = C[:, Ny // 2, :]  # Extract concentration slice at mid-plane (y = 500 m)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Z, C_slice, cmap='viridis', edgecolor='none')

# Labels and title
ax.set_xlabel("Crosswind Distance (m)")
ax.set_ylabel("Vertical Height (m)")
ax.set_zlabel("SO Concentration (g/m)")
ax.set_title("3D Chemically Reactive Power Plant Plume (y = 500 m)")

plt.show()
```
