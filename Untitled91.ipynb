```python
import numpy as np
import matplotlib.pyplot as plt

def box_muller(n):
    """Generate n pairs of independent standard normal random variables using Box-Muller Transform."""
    U1 = np.random.uniform(0, 1, n)
    U2 = np.random.uniform(0, 1, n)

    R = np.sqrt(-2 * np.log(U1))
    Theta = 2 * np.pi * U2

    X = R * np.cos(Theta)  # Standard normal variable
    Y = R * np.sin(Theta)  # Independent standard normal variable

    return X, Y

# Simulating hydrological discharge anomalies
n_samples = 5000  # Number of synthetic discharge samples
X, Y = box_muller(n_samples)

# Assume discharge anomalies follow N(0,1) and scale to realistic values
mean_discharge = 100  # Mean river discharge in m³/s
std_dev_discharge = 15  # Standard deviation in m³/s

discharge_anomalies = mean_discharge + std_dev_discharge * X  # Transforming standard normal to actual values

# Plot histogram of generated hydrological anomalies
plt.figure(figsize=(12,6))
plt.hist(discharge_anomalies, bins=50, density=True, alpha=0.7, color='b', label="Simulated Discharge Anomalies")
plt.axvline(mean_discharge, color='r', linestyle='dashed', linewidth=2, label="Mean Discharge")
plt.xlabel("River Discharge (m³/s)")
plt.ylabel("Density")
plt.title("Histogram of Simulated River Discharge Anomalies")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from scipy.linalg import cholesky

def generate_spatial_rainfall(grid_size, L, sigma):
    """
    Generate a spatially correlated Gaussian rainfall field using Cholesky decomposition.

    Parameters:
    grid_size (int): Number of grid points along one dimension (total grid = grid_size x grid_size).
    L (float): Correlation length (in km).
    sigma (float): Standard deviation of rainfall (in mm).

    Returns:
    RainfallField (2D ndarray): Simulated rainfall field.
    """
    # Step 1: Create spatial grid (Assume 1 km resolution)
    x = np.arange(grid_size)
    y = np.arange(grid_size)
    X, Y = np.meshgrid(x, y)
    coords = np.column_stack([X.ravel(), Y.ravel()])

    # Step 2: Compute pairwise distances and construct covariance matrix
    distances = squareform(pdist(coords))  # Compute Euclidean distances
    A_n = np.exp(-distances / L)  # Exponential correlation function

    # Step 3: Cholesky decomposition A_n = L L^T
    L_n = cholesky(A_n + 1e-6 * np.eye(A_n.shape[0]), lower=True)  # Add small value for stability

    # Step 4: Generate standard normal field
    Zn = np.random.randn(grid_size**2)

    # Step 5: Generate correlated rainfall field
    X = L_n @ Zn

    # Step 6: Transform to positive rainfall values (Lognormal transformation)
    RainfallField = np.exp(sigma * X).reshape(grid_size, grid_size)

    return RainfallField

# Example: Generate a spatial rainfall field (50x50 km region)
grid_size = 50  # 50x50 grid
correlation_length = 10  # Correlation decays over 10 km
rainfall_std = 1.0  # Lognormal scale

rainfall_field = generate_spatial_rainfall(grid_size, correlation_length, rainfall_std)

# Plot the generated rainfall field
plt.figure(figsize=(8, 6))
plt.imshow(rainfall_field, cmap='Blues', origin='lower', extent=[0, 50, 0, 50])
plt.colorbar(label="Rainfall (mm)")
plt.xlabel("X (km)")
plt.ylabel("Y (km)")
plt.title("Simulated Spatially Correlated Rainfall Field")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def scs_runoff(P, CN):
    """
    Compute runoff (Q) using the SCS Curve Number (CN) method.

    Parameters:
    P (float or array): Rainfall depth (mm)
    CN (float): Curve Number (ranges from 30-100)

    Returns:
    Q (float or array): Runoff depth (mm)
    """
    S = 25400 / CN - 254  # Potential maximum retention (mm)
    Ia = 0.2 * S  # Initial abstraction (mm)

    # Compute runoff Q
    Q = np.where(P > Ia, ((P - Ia)**2) / (P - Ia + S), 0)
    return Q

# Example: Compute runoff for different rainfall amounts
rainfall = np.linspace(0, 100, 100)  # Rainfall from 0 to 100 mm
CN = 75  # Example Curve Number for moderate urban area

runoff = scs_runoff(rainfall, CN)

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(rainfall, runoff, label=f'CN = {CN}', color='b')
plt.xlabel("Rainfall (mm)")
plt.ylabel("Runoff (mm)")
plt.title("Rainfall vs. Runoff using SCS Curve Number Method")
plt.legend()
plt.grid()
plt.show()
```

```python
def unit_hydrograph(Qr, Tp, duration=24, dt=1):
    """
    Compute streamflow hydrograph using a unit hydrograph approach.

    Parameters:
    Qr (float): Total runoff volume (mm)
    Tp (float): Time to peak (hours)
    duration (int): Total simulation time (hours)
    dt (int): Time step (hours)

    Returns:
    time (array): Time steps
    Q (array): Discharge (m³/s)
    """
    time = np.arange(0, duration, dt)  # Time array
    C = 1 / (Tp * np.exp(-1))  # Normalization constant

    # Unit Hydrograph equation
    Q = C * Qr * (time / Tp) * np.exp(-time / Tp)

    return time, Q

# Example: Compute streamflow for a runoff event
Qr = 50  # Runoff volume (mm)
Tp = 5   # Time to peak (hours)

time, discharge = unit_hydrograph(Qr, Tp)

# Plot hydrograph
plt.figure(figsize=(8, 5))
plt.plot(time, discharge, label="Streamflow Hydrograph", color='g')
plt.xlabel("Time (hours)")
plt.ylabel("Discharge (m³/s)")
plt.title("Unit Hydrograph-Based Streamflow Simulation")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def flow_duration_curve(flow_data):
    """
    Compute and plot a Flow Duration Curve (FDC).

    Parameters:
    flow_data (array): Streamflow time series (e.g., daily discharge in m³/s)
    """
    n = len(flow_data)
    sorted_flows = np.sort(flow_data)[::-1]  # Sort in descending order
    exceedance_prob = np.arange(1, n + 1) / (n + 1)  # Compute exceedance probability

    # Plot FDC
    plt.figure(figsize=(8, 5))
    plt.plot(exceedance_prob * 100, sorted_flows, marker="o", linestyle="-", color="b")
    plt.xscale("log")  # Log scale for better visualization
    plt.xlabel("Exceedance Probability (%)")
    plt.ylabel("Flow (m³/s)")
    plt.title("Flow Duration Curve")
    plt.grid(True, which="both", linestyle="--")
    plt.show()

# Example: Generate a synthetic flow dataset (or use real streamflow data)
np.random.seed(42)
synthetic_flow_data = np.random.gamma(shape=2, scale=10, size=365)  # Simulating daily flow

# Generate FDC
flow_duration_curve(synthetic_flow_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define transition matrix for water quality states
P = np.array([
    [0.6, 0.3, 0.1, 0.0],  # Good
    [0.2, 0.5, 0.2, 0.1],  # Moderate
    [0.1, 0.3, 0.4, 0.2],  # Poor
    [0.0, 0.1, 0.3, 0.6]   # Severe Pollution
])

states = ["Good", "Moderate", "Poor", "Severe Pollution"]

def simulate_water_quality(P, initial_state, num_steps):
    """ Simulates water quality transitions using a Markov chain """
    current_state = initial_state
    state_sequence = [states[current_state]]

    for _ in range(num_steps):
        current_state = np.random.choice(len(P), p=P[current_state])
        state_sequence.append(states[current_state])

    return state_sequence

# Simulate for 50 days starting from "Moderate" quality
num_steps = 50
initial_state = 1  # Moderate
sequence = simulate_water_quality(P, initial_state, num_steps)

# Convert to numerical states for plotting
state_indices = [states.index(s) for s in sequence]

# Plot the results
plt.figure(figsize=(10, 4))
plt.plot(state_indices, marker="o", linestyle="-", color="b")
plt.xlabel("Day")
plt.ylabel("Water Quality State")
plt.title("Water Quality Transitions Over Time")
plt.yticks(range(len(states)), states)  # Label y-axis with state names
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the target distribution (pollutant concentration model)
def target_pdf(C, M=10, V=100, lambda_=0.1, t=5):
    if C <= 0:  # Concentration cannot be negative
        return 0
    return np.exp(-((C - (M/V) * np.exp(-lambda_ * t))**2) / (2 * 0.5**2))

# Metropolis-Hastings Sampling
def metropolis_hastings(n_samples=10000, proposal_std=0.1):
    samples = []
    C_current = 1  # Initial concentration
    for _ in range(n_samples):
        # Propose a new state from a normal distribution
        C_proposed = np.random.normal(C_current, proposal_std)

        # Compute acceptance probability
        acceptance_ratio = target_pdf(C_proposed) / target_pdf(C_current)
        acceptance_probability = min(1, acceptance_ratio)

        # Accept or reject the new sample
        if np.random.uniform(0, 1) < acceptance_probability:
            C_current = C_proposed

        samples.append(C_current)

    return np.array(samples)

# Run Metropolis-Hastings algorithm
samples = metropolis_hastings()

# Plot the estimated distribution of pollutant concentration
plt.figure(figsize=(10, 5))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='b', label="Sampled Distribution")
plt.xlabel("Pollutant Concentration (mg/L)")
plt.ylabel("Density")
plt.title("Metropolis-Hastings Sampling of Pollutant Concentration")
plt.legend()
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the target pollutant concentration distribution (truncated normal)
def target_pdf(C, mean=2.0, std=0.5):
    if C <= 0:
        return 0
    return norm.pdf(C, mean, std)  # Normal distribution centered at 2 mg/L

# Define Metropolis-Hastings proposal distribution (Gaussian random walk)
def proposal_distribution(C, proposal_std=0.3):
    return np.random.normal(C, proposal_std)

# Compute acceptance probability
def acceptance_prob(C_current, C_proposed):
    return min(1, target_pdf(C_proposed) / target_pdf(C_current))

# Choose two arbitrary concentration states to check detailed balance
C1 = 1.5  # Initial pollutant concentration
C2 = 2.5  # Another concentration state

# Compute transition probabilities
q_C2_given_C1 = norm.pdf(C2, loc=C1, scale=0.3)  # Proposal density q(C2 | C1)
q_C1_given_C2 = norm.pdf(C1, loc=C2, scale=0.3)  # Proposal density q(C1 | C2)

alpha_C1_C2 = acceptance_prob(C1, C2)
alpha_C2_C1 = acceptance_prob(C2, C1)

# Compute P(C1 -> C2) and P(C2 -> C1)
P_C1_C2 = q_C2_given_C1 * alpha_C1_C2
P_C2_C1 = q_C1_given_C2 * alpha_C2_C1

# Verify detailed balance condition
lhs = target_pdf(C1) * P_C1_C2  # π(C1) * P(C1 -> C2)
rhs = target_pdf(C2) * P_C2_C1  # π(C2) * P(C2 -> C1)

# Print results
print(f"P(C1 -> C2) = {P_C1_C2:.5f}, P(C2 -> C1) = {P_C2_C1:.5f}")
print(f"LHS (π(C1) P(C1 -> C2)) = {lhs:.5f}")
print(f"RHS (π(C2) P(C2 -> C1)) = {rhs:.5f}")
print("Detailed Balance Condition Satisfied:", np.isclose(lhs, rhs, atol=1e-4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn, rand

# Simulation parameters
N = 5000   # Number of particles
T = 100    # Time steps
dt = 0.1   # Time step size
v_x, v_y = 0.05, 0.02  # Advection velocity
D = 0.02   # Diffusion coefficient
decay_rate = 0.001  # First-order decay rate

# Initialize particle positions randomly
x = np.zeros((N, 2))
x[:, 0] = np.random.uniform(-2, 2, N)  # X positions
x[:, 1] = np.random.uniform(-2, 2, N)  # Y positions

# Store trajectories (initialize with NaN to track disappearing particles)
trajectories = np.full((T, N, 2), np.nan)

# Time evolution
for t in range(T):
    # Store current positions
    trajectories[t, :, :] = x

    # Random diffusion step
    dx_diff = np.sqrt(2 * D * dt) * randn(N)
    dy_diff = np.sqrt(2 * D * dt) * randn(N)

    # Advection step
    dx_adv = v_x * dt
    dy_adv = v_y * dt

    # Update positions
    x[:, 0] += dx_adv + dx_diff
    x[:, 1] += dy_adv + dy_diff

    # Apply decay (set disappeared particles to NaN)
    survival_prob = np.exp(-decay_rate * t * dt)
    mask = rand(N) < survival_prob
    x[~mask] = np.nan  # Mark decayed particles as NaN

# Plot final state (ignore NaNs)
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], alpha=0.5, s=5, color='b')
plt.title("Contaminant Spread After {} Time Steps".format(T))
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.grid()
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data: Measured natural frequencies (Hz)
observed_frequencies = np.array([2.0, 2.1, 1.9, 2.05, 2.02])

# Known mass (kg)
m = 5.0

# Initialize Gibbs sampling parameters
num_samples = 5000
k_samples = np.zeros(num_samples)  # Stiffness samples
c_samples = np.zeros(num_samples)  # Damping coefficient samples

# Prior mean and standard deviations
mu_k, sigma_k = 1000, 100  # Prior for stiffness (N/m)
mu_c, sigma_c = 10, 2      # Prior for damping coefficient (Ns/m)

# Initial values
k_samples[0] = 1000
c_samples[0] = 10

# Gibbs Sampling Loop
for i in range(1, num_samples):
    c_current = c_samples[i - 1]

    # Compute mean and variance for k given c
    variance_k = sigma_k**2 / len(observed_frequencies)
    mean_k = (sum((2 * np.pi * observed_frequencies)**2 * m) + mu_k / sigma_k**2) * variance_k

    # Sample k from Normal Distribution
    k_samples[i] = np.random.normal(mean_k, np.sqrt(variance_k))

    k_current = k_samples[i]

    # Compute mean and variance for c given k
    variance_c = sigma_c**2 / len(observed_frequencies)
    mean_c = mu_c + (np.random.randn() * np.sqrt(variance_c))  # Sampling

    # Sample c from Normal Distribution
    c_samples[i] = np.random.normal(mean_c, np.sqrt(variance_c))

# Plot Results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(k_samples[1000:], bins=30, color="blue", alpha=0.7, density=True)
plt.title("Posterior Distribution of Stiffness (k)")
plt.xlabel("Stiffness (N/m)")
plt.ylabel("Probability Density")

plt.subplot(1, 2, 2)
plt.hist(c_samples[1000:], bins=30, color="red", alpha=0.7, density=True)
plt.title("Posterior Distribution of Damping Coefficient (c)")
plt.xlabel("Damping Coefficient (Ns/m)")
plt.ylabel("Probability Density")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given Parameters
L = 2.0       # Beam length (m)
P = 1000      # Applied force (N)
E = 200e9     # Young’s modulus (Pa)
rho = 7850    # Density (kg/m^3)
delta_max = 0.002  # Maximum allowable displacement (m)
sigma_noise = 0.0001  # Measurement noise standard deviation

# Simulated experimental displacement observations (actual measured values)
delta_obs = np.array([0.0018, 0.0019, 0.0021, 0.0020])

# Number of samples for Gibbs Sampling
num_samples = 5000

# Design variable initialization (Cross-sectional areas A1, A2)
A1_samples = np.zeros(num_samples)
A2_samples = np.zeros(num_samples)

# Prior distributions (Initial values)
A1_samples[0] = 0.01  # Initial area (m^2)
A2_samples[0] = 0.01  # Initial area (m^2)

# Limits for cross-sectional areas
A_min = 0.005  # Minimum area (m^2)
A_max = 0.05   # Maximum area (m^2)

# Likelihood function (probability of observed displacement given A1, A2)
def likelihood(A1, A2):
    I_eq = (A1**2 + A2**2) / 2
    delta_pred = (P * L**3) / (3 * E * I_eq)  # Predicted displacement
    return np.prod(norm.pdf(delta_obs, loc=delta_pred, scale=sigma_noise))

# Gibbs Sampling with Bayesian Updating
for i in range(1, num_samples):
    A2_current = A2_samples[i - 1]

    # Compute I_eq for current A2
    I_eq = (A1_samples[i - 1]**2 + A2_current**2) / 2

    # Compute required A1 to satisfy displacement constraint
    A1_new = ((P * L**3) / (3 * E * delta_max))**(1/2) if I_eq != 0 else A_min

    # Bayesian Updating: Modify A1 using Likelihood
    prior_A1 = norm.pdf(A1_new, loc=0.01, scale=0.005)  # Normal prior
    posterior_A1 = likelihood(A1_new, A2_current) * prior_A1

    # Sample A1 using Metropolis criterion
    if np.random.rand() < posterior_A1:
        A1_samples[i] = min(max(A1_new, A_min), A_max)
    else:
        A1_samples[i] = A1_samples[i - 1]

    A1_current = A1_samples[i]

    # Compute I_eq for current A1
    I_eq = (A1_current**2 + A2_samples[i - 1]**2) / 2

    # Compute required A2 to satisfy displacement constraint
    A2_new = ((P * L**3) / (3 * E * delta_max))**(1/2) if I_eq != 0 else A_min

    # Bayesian Updating: Modify A2 using Likelihood
    prior_A2 = norm.pdf(A2_new, loc=0.01, scale=0.005)  # Normal prior
    posterior_A2 = likelihood(A1_current, A2_new) * prior_A2

    # Sample A2 using Metropolis criterion
    if np.random.rand() < posterior_A2:
        A2_samples[i] = min(max(A2_new, A_min), A_max)
    else:
        A2_samples[i] = A2_samples[i - 1]

# Remove burn-in samples (first 1000)
A1_samples = A1_samples[1000:]
A2_samples = A2_samples[1000:]

# Plot Results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(A1_samples, bins=30, color="blue", alpha=0.7, density=True)
plt.title("Posterior Distribution of A1 (Updated)")
plt.xlabel("A1 (m^2)")
plt.ylabel("Probability Density")

plt.subplot(1, 2, 2)
plt.hist(A2_samples, bins=30, color="red", alpha=0.7, density=True)
plt.title("Posterior Distribution of A2 (Updated)")
plt.xlabel("A2 (m^2)")
plt.ylabel("Probability Density")

plt.tight_layout()
plt.show()

# Print the mean of the sampled values as optimal design areas
print(f"Updated A1: {np.mean(A1_samples):.5f} m^2")
print(f"Updated A2: {np.mean(A2_samples):.5f} m^2")
print(f"Updated Minimum Weight: {rho * (np.mean(A1_samples) + np.mean(A2_samples)) * L:.2f} kg")
```

```python
import numpy as np
from scipy.optimize import minimize

# Given parameters
l = 1.0       # Reference length
P = 1000.0    # Load force
E = 200e9     # Young’s modulus (e.g., for steel)
c0 = 0.01     # Compliance limit (arbitrary)

# Objective function (weight function in terms of x1, x2)
def objective(x):
    x1, x2 = x
    return (5 * l * P) / (E * x1) + (3 * l * P) / (E * x2)

# Constraint function (Compliance limit)
def compliance_constraint(x):
    x1, x2 = x
    return c0 - (1/x1 + 1/x2)  # Example form, needs proper derivation

# Define bounds (x1, x2 must be positive)
bounds = [(1e-6, None), (1e-6, None)]  # Prevent division by zero

# Initial guess
x0 = [0.01, 0.01]

# Define the constraint in Scipy format
constraint = {'type': 'ineq', 'fun': compliance_constraint}

# Solve the optimization problem
solution = minimize(objective, x0, method='SLSQP', constraints=[constraint], bounds=bounds)

# Extract results
optimal_x1, optimal_x2 = solution.x
optimal_A1 = P / (E * optimal_x1)
optimal_A2 = P / (E * optimal_x2)

print("Optimal nondimensional variables: x1 =", optimal_x1, ", x2 =", optimal_x2)
print("Optimal cross-sectional areas: A1 =", optimal_A1, "m^2, A2 =", optimal_A2, "m^2")
print("Minimum weight:", solution.fun)
```
